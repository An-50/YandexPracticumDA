# Описание данных

**Данные для первой части**

Файл **/datasets/hypothesis.csv**.

* **Hypothesis** — краткое описание гипотезы;
* **Reach** — охват пользователей по 10-балльной шкале;
* **Impact** — влияние на пользователей по 10-балльной шкале;
* **Confidence** — уверенность в гипотезе по 10-балльной шкале;
* **Efforts** — затраты ресурсов на проверку гипотезы по 10-балльной шкале. Чем больше значение Efforts, тем дороже проверка гипотезы.

**Данные для второй части**

Файл **/datasets/orders.csv**.
* **transactionId** — идентификатор заказа;
* **visitorId** — идентификатор пользователя, совершившего заказ;
* **date** — дата, когда был совершён заказ;
* **revenue** — выручка заказа;
* **group** — группа A/B-теста, в которую попал заказ.

Файл **/datasets/visitors.csv**.
* **date** — дата;
* **group** — группа A/B-теста;
* **visitors** — количество пользователей в указанную дату в указанной группе A/B-теста

```python
hypothesis
```

```python
hypothesis['ICE'] = hypothesis['Impact'] * hypothesis['Confidence'] / hypothesis['Efforts']
hypothesis[['Hypothesis', 'ICE']].sort_values('ICE', ascending=False).round(2)
```

### Примените фреймворк RICE для приоритизации гипотез. Отсортируйте их по убыванию приоритета.

```python
hypothesis['RICE'] = (hypothesis['Reach'] * hypothesis['Impact'] * hypothesis['Confidence']) / hypothesis['Efforts']
hypothesis[['Hypothesis', 'RICE']].sort_values('RICE', ascending=False)
```

Параметр RICE учитывает охват пользователей.

## Часть 2. Анализ A/B-теста

### Постройте график кумулятивной выручки по группам. Сделайте выводы и предположения.

```python
orders.info()
```

```python
orders.rename(columns={'transactionId':'transactionid', 'visitorId':'visitorid'}, inplace=True)
```

```python
orders.head()
```

```python
visitors.info()
```

```python
visitors.head()
```

```python
datesGroups = orders[['date','group']].drop_duplicates()
ordersAggregated = datesGroups.apply(lambda x:
                                        orders[np.logical_and(
                                        orders['date'] <= x['date'], orders['group'] == x['group'])]
                                        .agg({'date' : 'max', 
                                              'group' : 'max', 
                                              'visitorid' : 'nunique', 
                                              'transactionid' : 'nunique', 
                                              'revenue' : 'sum'}), axis=1)\
                                        .sort_values(by=['date','group'])
```

```python
ordersAggregated.head(6)
```

```python
visitorsAggregated = datesGroups.apply(lambda x: 
                                        visitors[np.logical_and(
                                        visitors['date'] <= x['date'], visitors['group'] == x['group'])]
                                        .agg({'date' : 'max', 
                                              'group' : 'max', 
                                              'visitors' : 'sum'}), axis=1)\
                                        .sort_values(by=['date','group']) 
visitorsAggregated.head(6)
```

```python
cumulativeData = ordersAggregated.merge(visitorsAggregated, left_on=['date', 'group'], right_on=['date', 'group'])
cumulativeData.columns = ['date', 'group', 'buyers', 'orders', 'revenue', 'visitors']
```

```python
cumulativeData.head(6)
```

```python
cumulativeRevenueA = cumulativeData[cumulativeData['group']=='A'][['date','revenue', 'orders']]
cumulativeRevenueB = cumulativeData[cumulativeData['group']=='B'][['date','revenue', 'orders']]

plt.figure(figsize=(16,8))
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue'], label='A')
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue'], label='B')
plt.xticks(rotation=45)
plt.title('Кумулятивная выручка по дням и группам')
plt.ylabel("Кумулятивная выручка")

plt.legend()
plt.grid(True)

plt.show();
```

Группа В лучше по показателю, чем группа А. Но в группе В есть резкий скачек, похожий на выброс.

### Постройте график кумулятивного среднего чека по группам. Сделайте выводы и предположения.

```python
plt.figure(figsize=(16,8))
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue'] / cumulativeRevenueA['orders'], label='A')
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue'] / cumulativeRevenueB['orders'], label='B')

plt.title('Кумулятивный средний чек по группам')
plt.ylabel("Выручка")
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

На графике тоже видно влияние аномальных значений на кумулятивный средний чек в группе В. 

### Постройте график относительного изменения кумулятивного среднего чека группы B к группе A. Сделайте выводы и предположения.


```python
merge_cum_rev = cumulativeRevenueA.merge(cumulativeRevenueB, on='date', suffixes=['A','B'])
merge_cum_rev.head()
```

```python
plt.figure(figsize=(16,8))
plt.plot(merge_cum_rev['date'] ,(
        (merge_cum_rev['revenueB'] / merge_cum_rev['ordersB']) / (merge_cum_rev['revenueA'] / merge_cum_rev['ordersA']) - 1))

plt.title('Относительное изменение кумулятивного среднего чека группы B к группе A')
plt.grid(True)
plt.xticks(rotation=45)
plt.axhline(y=0, color='red', linestyle='-')
plt.show();
```

На графике тоже видно влияние аномальных значений. Есть резкие перепады, как скачки, так и спады. После выброса группа В имеет средний чек выше, но разница постепенно идёт на спад.

### Постройте график кумулятивной конверсии по группам. Сделайте выводы и предположения.

```python
cumulativeData['conversion'] = cumulativeData['orders'] / cumulativeData['visitors']

cummulativeA = cumulativeData[cumulativeData['group']=='A'][['date', 'conversion']]
cummulativeB = cumulativeData[cumulativeData['group']=='B'][['date', 'conversion']]

plt.figure(figsize=(16,8))
plt.plot(cummulativeA['date'], cummulativeA['conversion'], label='A')
plt.plot(cummulativeB['date'], cummulativeB['conversion'], label='B')
plt.ylim([0.02, 0.04])

plt.grid(True)
plt.title('Кумулятивная конверсия по группам')
plt.ylabel("Выручка")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

В начале графика сильные колебания, но потом график сглаживался. Показатель в группе В стабильно больше.

### Постройте график относительного изменения кумулятивной конверсии группы B к группе A. Сделайте выводы и предположения.

```python
merge_cum_conv = cummulativeA.merge(cummulativeB, on='date', suffixes=['A','B'])
merge_cum_conv.head()
```

```python
plt.figure(figsize=(16,8))
plt.plot(merge_cum_conv['date'], merge_cum_conv['conversionB'] / merge_cum_conv['conversionA'] - 1)

plt.title('Относительное изменение конверсии')
plt.grid(True)
plt.xticks(rotation=45)
plt.ylim([-0.2, 0.3])
plt.axhline(y=0, color='red', linestyle='-');
```

В начале группа В уступала группе А по конверсии, но затем, большую часть теста имела показатели выше.

### Посчитайте статистическую значимость различий в конверсии между группами по «сырым» данным. Сделайте выводы и предположения.


```python
visitors_daily_A = visitors[visitors['group'] == 'A'][['date', 'visitors']]
visitors_daily_A.columns = ['date', 'visitorsPerDateA']
```

```python
visitors_daily_B = visitors[visitors['group'] == 'B'][['date', 'visitors']]
visitors_daily_B.columns = ['date', 'visitorsPerDateB']
```

```python
orders_countA = orders[orders['group'] == 'A'].groupby('visitorid').agg({'transactionid' : 'nunique'})['transactionid'].reset_index()
orders_countA.columns = ['userid', 'orders']
```

```python
orders_countB = orders[orders['group'] == 'B'].groupby('visitorid').agg({'transactionid' : 'nunique'})['transactionid'].reset_index()
orders_countB.columns = ['userid', 'orders']
```

```python
sampleA = pd.concat([orders_countA['orders'],
                     pd.Series(0, index = np.arange(visitors_daily_A['visitorsPerDateA'].sum() - len(orders_countA)), name='all_orders')], axis=0)

sampleB = pd.concat([orders_countB['orders'],
                     pd.Series(0, index = np.arange(visitors_daily_B['visitorsPerDateB'].sum() - len(orders_countB)), name='all_orders')], axis=0)
```

```python
sampleB.mean() / sampleA.mean()
```

Посчитаем статистическую значимость различий в конверсии.

**Нулевая гипотеза H0**: Конверсия в группе A **равна** конверсии в группе B.<br> 
**Альтернативная гипотеза H1**: Конверсия в группе A **не равна** конверсии в группе B.

```python
st_con1 = st.mannwhitneyu(sampleA, sampleB, alternative = 'two-sided')[1]
print('{0:.3f}'.format(st_con1))
if st_con1 > 0.05:
    print('Не получилось отвергнуть нулевую гипотезу')
else:
    print('Отвергаем нулевую гипотезу')
```

Отвергаем нулевую гипотезу. Различия в конверсии есть

### Посчитайте статистическую значимость различий в среднем чеке заказа между группами по «сырым» данным. Сделайте выводы и предположения.


**Нулевая гипотеза H0**: различий в среднем чеке между группами нет.<br> 
**Альтернативная гипотеза H1**: различия в среднем чеке между группами есть.

```python
st_rev1 = st.mannwhitneyu(orders[orders['group'] == 'A']['revenue'], orders[orders['group'] == 'B']['revenue'],
                          alternative = 'two-sided')[1]
print('{0:.3f}'.format(st_rev1))
if st_rev1 > 0.05:
    print('Не получилось отвергнуть нулевую гипотезу')
else:
    print('Отвергаем нулевую гипотезу')
```

```python
orders[orders['group'] == 'B']['revenue'].mean() / orders[orders['group'] == 'A']['revenue'].mean()-1
```

Не получилось отвергнуть нулевую гипотезу. Разница по среднему чеку между группами составила 25.9%

### Посчитайте статистическую значимость различий в конверсии между группами по «очищенным» данным. Сделайте выводы и предположения.


Сделаем срез аномальных пользователей: тех, кто оформил более 4 заказов и кто потратил более 58233.

```python
users_many_orders = pd.concat(
    [
     orders_countA[orders_countA['orders'] > 4]['userid'], 
     orders_countB[orders_countB['orders'] > 4]['userid']],
    axis=0)
users_expensive_orders = orders[orders['revenue'] > 58233]['visitorid']
users_expensive_orders.name = 'userid'
```

Создадим общую таблицу аномальных пользователей.

```python
abnormal_users = pd.concat(
    [    users_many_orders, users_expensive_orders],
    axis=0).drop_duplicates().sort_values()
print(abnormal_users.head())
len(abnormal_users)
```

Всего 15 аномальных пользователей.

```python
sampleA_filtered = pd.concat([orders_countA[np.logical_not(orders_countA['userid'].isin(abnormal_users))]['orders'],
                     pd.Series(0,
                               index = np.arange(visitors_daily_A['visitorsPerDateA'].sum() - len(orders_countA)),
                               name='all_orders')
                    ], axis=0)

sampleB_filtered = pd.concat([orders_countB[np.logical_not(orders_countB['userid'].isin(abnormal_users))]['orders'],
                     pd.Series(0,
                               index = np.arange(visitors_daily_B['visitorsPerDateB'].sum() - len(orders_countB)),
                               name='all_orders')
                    ], axis=0)
```

Посчитаем статистическую значимость различий в конверсии между группами.

**Нулевая гипотеза H0**: различий в среднем чеке между группами нет.<br> 
**Альтернативная гипотеза H1**: различия в среднем чеке между группами есть.

```python
st_con2 = st.mannwhitneyu(sampleA_filtered, sampleB_filtered, alternative = 'two-sided')[1]
print('{0:.3f}'.format(st_con2))
if st_con2 > 0.05:
    print('Не получилось отвергнуть нулевую гипотезу')
else:
    print('Отвергаем нулевую гипотезу')
```

```python
sampleB_filtered.mean() / sampleA_filtered.mean()-1
```

Отвергаем нулевую гипотезу.<br>
Различия по конверсии сохранились, прирост средней конверсии в группе В составил 15%
