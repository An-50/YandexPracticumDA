## Предобработка данных

удаляем данные с которыми точно не будем работать

```python
data = data[['locality_name','total_area','living_area','kitchen_area','rooms','ceiling_height','balcony','floor','floors_total','last_price','first_day_exposition','days_exposition','is_apartment','studio','open_plan','cityCenters_nearest']]

data.info()
```

Не указано число балконов — скорее всего, их нет. Меняем на 0. и меняем тип данных на целочисленный

```python
data['balcony'] = data['balcony'].fillna(value=0)
data['balcony'] = data['balcony'].astype('int')
```

## Расчёты и добавление результатов в таблицу

Добавляем расстояние в километрах

```python
data['cityCenters_nearest_km'] = (data['cityCenters_nearest'] / 1000).round()
```

Добавляем цену квадратного метра

```python
data['price_per_meter'] = (data['last_price'] / data['total_area']).astype('int')
```

Переводим цену квартиры в тысячи рублей

```python
data['last_price'] = (data['last_price'] / 1000).astype('int')
```

Добавляем день недели, месяц и год публикации объявления

```python
data['day'] = data['first_day_exposition'].dt.day_name()
data['month'] = data['first_day_exposition'].dt.month
data['year'] = data['first_day_exposition'].dt.year
```

Добавляем этаж квартиры; варианты — первый, последний, другой;

```python
def floor_type(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    elif row['floor'] > row['floors_total']:
        return 'мансарда'
    else:
        return 'другой'
    
data['floor_type'] = data.apply(floor_type, axis=1)
```

Проверяем ситуацию, что этаж рассматриваемой квартиры выше общего их числа в доме.

```python
(data['floor_type'] == 'мансарда').sum()
```

Этаж рассматриваемой квартиры выше общего их числа в доме отсутствуют.

Расчитываем соотношение жилой и общей площади, а также отношение площади кухни к общей

```python
data['living_area_ratio'] = data['living_area'] / data['total_area']
data['kitchen_area_ratio'] = data['kitchen_area'] / data['total_area']
```
