#### Рассчитаем Retention Rate, применяя когортный анализ.

```python
first_session_date = visits.groupby(['uid'])['session_date'].min()
first_session_date.name = 'first_session_date'
visits = visits.join(first_session_date,on='uid')
```

```python
visits['first_month'] = visits['first_session_date'].astype('datetime64[M]')
visits['visit_month'] = visits['session_date'].astype('datetime64[M]')
```

```python
visits['lifetime'] = (visits['visit_month'] - visits['first_month']) / np.timedelta64(1, 'M')
visits['lifetime'] = visits['lifetime'].round().astype('int')
```

```python
cohorts = visits.groupby(['first_month', 'lifetime']).agg({'uid': 'nunique'}).reset_index()
```

```python
initial_users_count = cohorts[cohorts['lifetime'] == 0][['first_month', 'uid']]
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'})
```

```python
cohorts = cohorts.merge(initial_users_count,on='first_month')
cohorts['retention'] = cohorts['uid']/cohorts['cohort_users']
```

```python
cohorts['first_month_str'] = cohorts['first_month'].dt.strftime('%Y-%m-%d')
retention_pivot = cohorts.pivot_table(
    index='first_month_str',
    columns='lifetime',
    values='retention',
    aggfunc='sum',
)
```

```python
sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('Когортный анализ')
sns.heatmap(retention_pivot, annot=True, fmt='.2%', linewidths=1, linecolor='gray', cmap='coolwarm')
plt.xlabel('Месяц активности')
plt.ylabel('Когорты')

plt.show();
```

```python
retention_pivot[1].mean() * 100 
```

- в основном, 92-96% пользователей уходят после первого месяца
- на второй месяц остается в среднем 6,5% пользователей

#### Рассчитаем среднее количество покупок на одного покупателя за определённый период, например за 6 месяцев

#### Выясним, как меняется LTV на покупателя по когортам

Помните, что LTV — накопительная метрика. Рассчитайте средний LTV по когортам за 6 месяцев; в расчёт включайте когорты, «прожившие» не менее 6 месяцев. Маржинальность сервиса — 100%. Отразите изменения метрики во времени на графике;

```python
first_orders = orders.groupby('uid').agg({'purchase_month': 'min'}).reset_index()
first_orders.columns = ['uid', 'first_order_month']


cohort_sizes = (
    first_orders.groupby('first_order_month')
    .agg({'uid': 'nunique'})
    .reset_index()
)
cohort_sizes.columns = ['first_month', 'n_buyers']
```

```python
cohorts = (
    orders.groupby(['first_month', 'purchase_month'])
    .agg({'revenue': 'sum'})
    .reset_index()
)

report = pd.merge(cohort_sizes, cohorts, on='first_month')
```

```python
# Маржинальность сервиса — 100%
margin_rate = 1

report['gp'] = report['revenue'] * margin_rate
report['age'] = (report['purchase_month'] - report['first_month']) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')
report['ltv'] = report['gp'] / report['n_buyers']

ltv = report.pivot_table(
    index='first_month', columns='age', values='ltv', aggfunc='mean'
).cumsum(axis=1).round(2)
```

```python
sns.set(style='white')
plt.figure(figsize=(15, 9))
plt.title('LTV')
ax = sns.heatmap(ltv, annot=True, fmt='.2f', linewidths=1, linecolor='gray', cmap='coolwarm')
ax.set_yticklabels(ltv.index.date)
plt.xlabel('Месяц активности')
plt.ylabel('Когорты')
plt.show();
```

```python
ltv6 = ltv.iloc[:7,:6]
```

```python
sns.set(style='white')
plt.figure(figsize=(9, 6))
plt.title('LTV')
ax = sns.heatmap(ltv6, annot=True, fmt='.2f', linewidths=1, linecolor='gray', cmap='coolwarm')
ax.set_yticklabels(ltv6.index.date)
plt.xlabel('Месяц активности')
plt.ylabel('Когорты')
plt.show();
```
